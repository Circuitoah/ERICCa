## ERICCa : Too Long Didn't Read

If you want to calculate the reaction cross section without the tutorial, copy this code and adjust where you see fit.

```python
#load in Package
import ERICCa as ERICCa

# Density: fit a two-point Fermi to 12C rms matter radius
my_12C_rho = ERICCa.Density()
A_12C = 12        # mass number
rms_12C = 2.32    # rms matter radius [fm]
my_12C_rho.rho_m_2pt_fermi(A_12C, rms_12C)

# Profile function from tabulated NN parameters at E=300 MeV
profile_funct = ERICCa.ProfileFunction(model_type="matter", E=300)

# CrossSection: project rho(r) -> rho(b) then compute sigma_R
cross_section = ERICCa.CrossSection()
r_mesh = np.linspace(0.001, 20, 100)
C_rho = cross_section.dens_b_interpolator(r_mesh, my_12C_rho.rho_m(r_mesh))

cross_section.sigma_R_matter(C_rho, rho_p=C_rho, Gamma=profile_funct.Gamma, Model="MOL")
```
